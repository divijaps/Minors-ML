```python
#Extracting the months of the entire dates by bringing it in desired format and making a data frame out of it
import datetime
dates =df_issue['Patent Expiration Date']
  
frame = pd.to_datetime(dates, dayfirst = True)
frame = pd.DataFrame([frame]).transpose()
frame['date']= frame
frame['Expiration month']= frame['date'].dt.month 
frame['Center']=df_issue['Center']
frame.drop('date', axis = 1, inplace = True)

frame
```

```python
#Adding the columns to the already existing data frame
df_corr['Expiration month']=frame['Expiration month']
df_corr['Patent Expiration Date']=frame['Patent Expiration Date']
df_corr
```
